## 📌 잠깐! SW 토막 상식

SW 업계에서는 줄 바꿈을 다음과 같이 부릅니다:

* 영어로는
  * line break
  * 또는 new line
  * 또는 line feed
* 한국어로는
  * 개행

또한 업계 표준으로, 컴퓨터는 줄 바꿈을 "\n"으로 인지합니다.

앞으로 강의에서는 "\n"을 줄 바꿈 대신 line break라고 표현하겠습니다.

## 알려줄게요! 문자열에서 line break를 제거하는 방법

문자열에서는 특정 문자를 [.replace(어쩌고, 저쩌고)](https://docs.python.org/3/library/stdtypes.html#str.replace)를 이용해서 대체할 수 있다.
  * 첫 번째 파라미터인 `어쩌고`에는 없애고 싶은 문자를 넣는다.
  * 두 번째 파라미터인 `저쩌고`에는 없앤 문자 자리에 대체할 문자를 넣는다.
    * 문자를 다른 문자로 대체하지 않고, 아예 제거하고 싶다면 이 자리에 공백 문자("" 또는 ''로 표현)를 넣는다.


## 예시

### 예시 1: 문자열에서 특정 문자를 대체하기

다음은 문자열 중 `"Melon"`을 `"Grape"`로 대체하는 코드입니다.

왼쪽에 위치한 화살표 버튼을 클릭하여 코드를 실행해보세요.

````python
fruits = "Apple Banana Melon Apple Melon Banana"

# .replace 를 사용해서 Melon을 Grape로 대체
fruits_replaced = fruits.replace("Melon", "Grape")

# 결과 출력
print(fruits_replaced)
````

### 예시 2: 문자열에서 특정 문자를 없애기

다음은 문자열 중 `"Melon"`만 제거하는 코드입니다.

왼쪽에 위치한 화살표 버튼을 클릭하여 코드를 실행해보세요.


````python
fruits = "Apple Banana Melon Apple Melon Banana"

# .replace를 사용해서 Melon을 제거
fruits_without_melon = fruits.replace("Melon", "")

print(fruits_without_melon)
````
