# STEP 3. 빨간색을 선택했는지 파악하기

어떤 사람이 빨간색을 선택했다면 이 사람의 정보는 빨간색 시트에, 초록색을 선택했다면 잇 사람의 정보는 초록색 시트에 넣어주어야 합니다.

그렇다면, 빨간색을 선택했는지 아닌지는 어떻게 알 수 있을까요?

## 알려줄게요!: 문자열에 어떤 값이 들어있는지 아닌지 알아내는 방법

`in` 키워드를 사용하면 문자열에 어떤 값이 들어있는지 아닌지 알아낼 수 있습니다.


`어쩌고 in 저쩌고`를 사용하면

* `어쩌고`가 `저쩌고`에 들어있다면 이 값은 True로 감정됩니다.
* `어쩌고`가 `저쩌고`에 들어있지 않다면 이 값은 False로 감정됩니다.

참고: `어쩌고 not in 저쩌고`를 사용하면 반대의 결과물이 나옵니다.

## 예시

### 예시 1: 문자열에 어떤 값이 들어있는지 확인하기

다음은 문자열에 Banana가 들어있는지 판단하는 코드입니다.

왼쪽에 있는 화살표 버튼을 클릭하여 코드를 실행해보세요.

````python
# 주어진 문장
fruits = "Apple Banana Melon Apple Melon Banana"

print("Banana" in fruits)
````

### 예시 2: 문자열에 어떤 값이 **안** 들어 있는지 확인하기

다음은 문자열에 Banana가 안 들어 있는지 판단하는 코드입니다.

왼쪽에 있는 화살표 버튼을 클릭하여 코드를 실행해보세요.

````python
print("Banana" not in fruits)
````

# STEP 4: 조건 X를만족하면 Y하기

## 알려줄게요!: If 문

컴퓨터가 어떤 조건 X를 만족할 때만 Y를 하게 만들려면, If 문(조건문)을 사용해야 합니다.

`어떤조건`을 만족하면 `어떤일`을 시키는 If 문은 다음과 같이 작성합니다.

```python
if 어떤조건:
  어떤일
```

## 예시

### 예시 1: 리스트에서 "e"가 있는 원소만 출력하기

````python
fruits = [
    "Apple",
    "Banana",
    "Melon",
    "Apple",
    "Melon",
    "Banana",
]

for fruit in fruits:
  if "e" in fruit:
    print(fruit)
````
